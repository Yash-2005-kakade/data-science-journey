# Student Digital Learning Behavior Analysis

## 1. Problem Statement

- With the rapid growth of digital learning platforms, students increasingly rely on smartphones, laptops, and online resources for education. However, excessive screen time and social media usage may affect academic performance and productivity.

- This project aims to explore how study habits, digital usage patterns, sleep duration, and attendance influence academic performance using Exploratory Data Analysis techniques.

## 2. Objectives

- Analyze the distribution of student learning behavior.

- Identify relationships between study hours and exam scores.

- Investigate the impact of social media usage on academic performance.

- Examine the effect of sleep patterns on productivity.

- Identify outliers and unusual patterns in the dataset.

- Perform correlation analysis between variables.

- Generate meaningful insights using data visualization techniques.

## 3. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 4. Load Dataset

```python
df = pd.read_csv("student_learning_behavior_dataset.csv")
df.head(10)
```

## 5. Dataset Overview

```python
df.info()
```

## 6. Statistical Summary

```python
df.describe()
```

## 7. Missing Value Analysis

```python
df.isnull().sum()
```

## 8. Distribution of Numerical Variables

```python
df.hist(figsize=(12,8))
plt.show()
```

### Insight

#### Histograms show how values are distributed across different variables such as:

- Study hours

- Screen time

- Exam scores

## 9. Study Hours vs Exam Score

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Study_Hours_Per_Day', y='Exam_Score', data=df)
plt.title("Study Hours vs Exam Score")
plt.show()
```

Insight

- A positive relationship exists between study hours and exam performance.

## 10. Social Media Usage Analysis

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Social_Media_Hours', y='Exam_Score', data=df)
plt.title("Social Media Usage vs Exam Score")
plt.show()
```

Insight

- Students spending more time on social media tend to have lower academic scores.

## 11. Sleep Hours Analysis

```python
plt.figure(figsize=(8,5))
sns.violinplot(x=df['Sleep_Hours'])
plt.title("Sleep Hours Distribution")
plt.show()
```

Insight

- Most students sleep between 6-8 hours, which is considered optimal for cognitive performance.

## 12. Attendance vs Exam Score

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Attendance_Percentage', y='Exam_Score', data=df)
plt.title("Attendance vs Exam Score")
plt.show()
```

Insight

- Higher attendance percentage strongly correlates with higher exam scores.

## 13. Feature Engineering

Creating a new feature called Productivity Index.

```python
df['Productivity_Index'] = (
    df['Study_Hours_Per_Day'] +
    df['Attendance_Percentage']/20 +
    df['Online_Courses_Completed']
)
```

Insight

- This index combines study behavior and attendance to estimate student productivity.

## 14.Outlier Detection

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df[['Study_Hours_Per_Day','Social_Media_Hours','Screen_Time_Hours']])
plt.show()
```

Insight

Outliers represent students with unusually high screen time or study hours.

## 15. Average Behavior Analysis

```python
df.mean()
```

## 16.Average Exam Score by Study Hours

```python
df['Study_Category'] = pd.cut(
    df['Study_Hours_Per_Day'],
    bins=[0,2,4,6,8],
    labels=['Low','Medium','High','Very High']
)

study_group = df.groupby('Study_Category')['Exam_Score'].mean()

print(study_group)
```

## 17. Key Findings

i) Students studying more than 5 hours daily achieve higher scores.

 ii) High social media usage negatively affects performance.

 iii) Balanced sleep patterns improve academic productivity.

 iv) Attendance percentage strongly influences exam results.

 v) Students completing more online courses demonstrate better learning outcomes.

## 18. Conclusion

The exploratory analysis of student digital learning behavior reveals that academic success is influenced by a combination of study habits, attendance, and lifestyle patterns.

Students who maintain balanced screen time, sufficient sleep, higher study hours, and consistent attendance tend to perform better academically. Conversely, excessive social media usage may negatively impact academic performance.
