# University Academic Performance Analysis Using NumPy

## 📄 Problem Statement
- Modern universities generate large volumes of academic data across multiple subjects and students. Efficient numerical analysis of this structured data is essential for understanding student performance trends, identifying academic strengths and weaknesses, and supporting data-driven decision-making.

- In this project, a dataset representing the marks of 500 students across 8 subjects is analyzed using NumPy 2D arrays. The goal is to perform structured academic performance evaluation through numerical computation and matrix-based operations.

## Objectives of the Study

The primary objective of this project is to analyze large-scale academic performance data using NumPy 2D arrays and numerical computing techniques.

The specific objectives of this study are:

- To structure student academic records into a two-dimensional matrix format.
- To examine overall and subject-wise performance trends across 500 students.
- To compute total marks and percentage scores for each student.
- To evaluate average performance in each subject across the university.
- To measure the variation and distribution of marks using statistical analysis.
- To identify high-performing students and academically at-risk students.
- To apply uniform grace marks using broadcasting techniques.
- To demonstrate array reshaping, transformation, and concatenation operations.
- To perform logical filtering for academic classification.
- To strengthen understanding of NumPy fundamentals including indexing, slicing, and matrix operations.

This project aims to build a strong numerical computing foundation required for advanced data science and machine learning applications.

### Importing numpy library

```python
import numpy as np
```

## Step 1: Creation of Academic Performance Dataset

The first step involves generating a structured academic dataset 
representing marks of 500 students across 8 subjects.

Each row represents an individual student.
Each column represents a subject.

All marks are generated within the range of 0 to 100.

```python
np.random.seed(42)

results = np.random.randint(0, 101, size=(500, 8))
results.shape
```

## Step 2: Examination of Dataset Structure and Properties

The dataset is examined to understand its size, dimensionality, 
and storage characteristics before performing further analysis.

### Total number of students

```python
print("Number of Students:", results.shape[0])
```

### Total subjects

```python
print("Number of Subjects:", results.shape[1])
```

### Total number of records

```python
print("Total Records:", results.size)
```

### Number of Dimensions

```python
print("Number of Dimensions:", results.ndim)
```

### DataType  of elements store in Array

```python
print("Data Type:", results.dtype)
```

## Step 3: Accessing Individual and Subject-Wise Records

In this step, we retrieve:
- Marks obtained by the first student
- Marks scored in a particular subject by all students
- Records of the first ten students

### Marks obtained by the first student

```python
print("Marks of First Student:", results[0])
```

### Marks scored in a particular subject by all students

```python
print("Marks in Subject 3:", results[:, 2])
```

### Records of the first ten students

```python
print("First 10 Students:\n", results[:10])
```

## Step 4: Calculation of Total Marks Obtained by Each Student

Total marks across all subjects are computed to evaluate 
overall student performance.

```python
total_marks = np.sum(results, axis=1)
total_marks[:10]
```

## Step 5: Determining Average Performance in Each Subject

The average marks for each subject are calculated 
to understand subject difficulty and scoring trends.

```python
subject_average = np.mean(results, axis=0)
subject_average
```

## Step 6: Identifying Highest and Lowest Overall Scores

This step identifies the maximum and minimum total marks 
obtained within the academic batch.

```python
print("max marks:",np.max(total_marks))

print("min marks:",np.min(total_marks))
```

## Step 7: Measuring Variation and Dispersion in Marks

Statistical measures are applied to understand consistency 
and variability in student performance.

### i) Overall Average Marks

```python
print("Overall Average Marks:", np.mean(results))
```

### ii) Standard Deviation

```python
print("Standard Deviation:", np.std(results))
```

### iii) Variance

```python
print("Variance:", np.var(results))
```

## Step 8: Applying Grace Marks Uniformly Across Dataset

Five grace marks are added to every subject for every student 
using broadcasting techniques.

```python
updated_results = results + 5
updated_results[:5]



```

## Step 9: Converting Total Marks into Percentage Scores

The overall percentage for each student is calculated 
based on maximum possible marks (800).

```python
percentage = (total_marks / 800) * 100
percentage[:10]
```

## Step 10: Identifying Distinction and At-Risk Students

Students scoring above 75% are categorized under distinction.
Subjects scoring below 35 marks are considered failures.

```python
distinction_count = np.sum(percentage > 75)
failure_count = np.sum(results < 35)

print(distinction_count)
print( failure_count)
```

## Step 11: Restructuring the Dataset into an Alternative Matrix Format

In order to demonstrate structural flexibility of numerical datasets,
the academic matrix is reshaped into a new two-dimensional structure
without altering the total number of stored marks.

This transformation validates understanding of reshape operations.

```python
reshaped_results = results.reshape(1000, 4)
reshaped_results.shape
```

## Step 12: Transforming the Academic Matrix into a Linear Sequence

The two-dimensional academic dataset is converted into a 
single continuous array containing all recorded marks.

This operation demonstrates flattening of structured data.

```python
flattened_results = results.flatten()
print(flattened_results.shape)
print(flattened_results)
```

## Step 13: Merging Two Academic Batches into a Unified Dataset

A second batch of 500 students is generated and merged 
with the original dataset to simulate institutional growth.

The merging is performed vertically to increase the student count.

```python
batch_two = np.random.randint(0, 101, size=(500, 8))

combined_batch = np.vstack((results, batch_two))
combined_batch.shape
```

## Step 14: Adding an Additional Subject to the Academic Structure

An extra subject is introduced into the existing dataset.
The new subject marks are appended horizontally, thereby
increasing the subject dimension.

```python
new_subject = np.random.randint(0, 101, size=(500, 1))

extended_dataset = np.hstack((results, new_subject))
extended_dataset.shape
```

## Step 15: Dividing Students into Two Equal Academic Groups

For comparative performance analysis,
the dataset is split into two equal halves.

This demonstrates splitting of 2D arrays.

```python
group_one, group_two = np.vsplit(results, 2)

group_one.shape, group_two.shape
```

## Step 16: Selecting a Random Sample of Students for Performance Audit

A subset of students is randomly selected
to simulate an academic performance audit.

This demonstrates usage of NumPy's random module.

```python
sample_indices = np.random.choice(500, size=20, replace=False)
results[sample_indices]
```

## Step 17: Evaluating Median Performance of the Academic Dataset

The median value is computed to identify 
the central performance level in the institution.

```python
print(np.median(results))
```

## Step 18: Sorting All Recorded Marks for Distribution Analysis

All marks are arranged in ascending order 
to study the distribution pattern.

```python
sorted_marks = np.sort(results, axis=None)
sorted_marks[:20]
```

## Step 19: Determining Unique Score Values Present in the Dataset

This step identifies how many distinct mark values 
exist within the academic dataset.

```python
unique_scores = np.unique(results)
len(unique_scores)
```
