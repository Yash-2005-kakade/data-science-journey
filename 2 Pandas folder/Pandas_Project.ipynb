## 7.Filtering Data

### i) Filter only movies:

```python
movies = df[df["type"] == "Movie"]
```

### ii) Filter TV shows:

```python
tv_shows = df[df["type"] == "TV Show"]
```

### iii) Filter content released after 2015:

```python
recent_content = df[df["release_year"] > 2015]
```

## 8.Sorting Data

### i)Sort by newest content:

```python
df.sort_values(by="release_year", ascending=False).head(10)
```

### ii) Sort by highest views:

```python
df.sort_values(by="views_millions", ascending=False).head(10)
```

## 9. Content Distribution

### i) Count Movies vs TV Shows:

```python
df["type"].value_counts()
```

## 10. Country Analysis

### i) Top countries producing content:

```python
df["country"].value_counts().head(10)
```

### ii) Group by country:

```python
df.groupby("country")["title"].count().sort_values(ascending=False)
```

## 11.Genre Analysis

### i) Most popular genres:

```python
df["genre"].value_counts()
```

## 12. Time-Based Analysis

### i) Convert date column to datetime: 

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

### ii) Extract year added:

```python
df["year_added"] = df["date_added"].dt.year
```

### iii) Content added per year:

```python
df["year_added"].value_counts().sort_index()
```

## 13. GroupBy Aggregation

### i) Average views by genre:

```python
df.groupby("genre")["views_millions"].mean()
```

### ii) Average views by country:

```python
df.groupby("country")["views_millions"].mean()
```

### iii) Content count by release year:

```python
df.groupby("release_year").size()
```

## 14. Pivot Table Analysis

### Create pivot table:

```python
pd.pivot_table(
    df,
    index="type",
    values="views_millions",
    aggfunc="mean"
)
```
