# Cricket Match Data Visualization using Matplotlib and Seaborn

The main challenge is to analyze and interpret this cricket data effectively. Using data visualization techniques helps convert complex datasets into understandable graphs and charts. This project uses visualization tools to identify patterns and trends in cricket match performance.

## 2.Objectives of the Project


The main objectives of this project are:

- To explore and analyze cricket match data.
- To visualize important match statistics using **Matplotlib and Seaborn**.
- To understand team performance based on runs scored and wickets lost.
- To analyze the influence of venues on match scores.
- To identify players who frequently receive **Player of the Match** awards.
- To extract meaningful insights from cricket match statistics.

## 3. Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 4. Loading and Exploring the Dataset

```python
df = pd.read_csv("cricket_dataset_600_rows.csv")

df.head(7)
```

```python
df.info()
```

### 4.2 Statistical Summary

```python
df.describe()
```

```python
df.isnull().sum()
```

## 5.Data Visualization and Analysis

### 5.1 Distribution of Runs Scored

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Runs_Scored'], bins=20, kde=True)
plt.title("Distribution of Runs Scored")
plt.xlabel("Runs Scored")
plt.ylabel("Frequency")
plt.show()
```

#### INSIGHT

- This histogram shows how runs are distributed across matches.
- Most teams score between 150 and 200 runs, indicating that this range is the most common scoring pattern in the dataset.
- Very high scores above 210 runs are less frequent.

### 5.2 Team-wise Match Count

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Team', data=df)
plt.title("Number of Matches Played by Each Team")
plt.xticks(rotation=45)
plt.show()
```

#### INSIGHT

- This chart shows how many matches each team played in the dataset.
- Teams such as India, Australia, and England appear frequently, indicating higher representation in the dataset.

### 5.3 Runs Scored by Each Team

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Team', y='Runs_Scored', data=df)
plt.title("Runs Distribution by Team")
plt.xticks(rotation=45)
plt.show()
```

#### INSIGHT

- The boxplot shows the spread of runs scored by different teams.
- Some teams have higher median scores, indicating stronger batting performance.
- The presence of outliers indicates occasional very high scoring matches.

### 5.4 Venue-wise Average Runs

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Venue', y='Runs_Scored', data=df)
plt.title("Average Runs Scored at Different Venues")
plt.xticks(rotation=45)
plt.show()
```

#### INSIGHT

- This visualization shows how the average runs vary across different venues.
- Some stadiums appear to favor high scoring matches, while others show lower average runs, possibly due to pitch conditions.

### 5.5 Runs vs Wickets Relationship

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Runs_Scored', y='Wickets_Lost', data=df)
plt.title("Runs Scored vs Wickets Lost")
plt.show()
```

#### INSIGHT

- This scatter plot shows the relationship between runs scored and wickets lost.
- Teams that lose fewer wickets often score higher runs, indicating stable batting performance.

### 5.6 Match Result Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Result', data=df)
plt.title("Match Result Distribution")
plt.show()
```

#### INSIGHT

- This plot shows the number of matches won and lost in the dataset.
- A balanced distribution of wins and losses indicates competitive matches.

### 5.7 Player of the Match Frequency

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Player_of_Match', data=df)
plt.title("Player of the Match Awards")
plt.xticks(rotation=45)
plt.show()
```

#### INSIGHT

- This visualization shows how frequently players receive the Player of the Match award.
- Players like Virat Kohli, Joe Root, and Steve Smith appear frequently, indicating consistent performance.

### 5.8 Correlation Heatmap

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## 6. Key Insights

- Most teams score between 150 and 200 runs in matches.

- Certain venues tend to produce higher scoring matches.

- Teams that lose fewer wickets generally score more runs.

- Some players consistently perform well and receive Player of the Match awards.

- The dataset shows a relatively balanced number of wins and losses.
